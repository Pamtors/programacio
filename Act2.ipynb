```python
costat1 = float(input("Introdueix la distancia del primer costat: "))
costat2 = float(input("Introdueix la distancia del segon costat: "))

operacio = input("Quina operació vols fer? (A - àrea, P - perímetre): ")


operacio = operacio.upper()

if operacio == "A":
    area = costat1 * costat2
    print(f"L'àrea del rectangle és: {area}")

elif operacio == "P":
    perimetre = 2 * (costat1 + costat2)
    print(f"El perímetre del rectangle és: {perimetre}")

else:
    print("Error: operació no vàlida. Has d'introduir 'A' per àrea o 'P' per perímetre.")
    
```
