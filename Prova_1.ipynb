```python
# Variables de pes
pes_examen = 0.6
pes_exercicis = 0.4

# Nota mínima per aprovar
nota_minima_aprovar = 5.0

# Nota mínima a l'examen per poder fer mitja
nota_examen = 3.5

# Càlcul del que aporta l'examen
aportacio_examen = nota_examen * pes_examen

# Càlcul de la nota necessària als exercicis
nota_necessaria_exercicis = (nota_minima_aprovar - aportacio_examen) / pes_exercicis

# Resultat
print("Per aprovar amb un", nota_examen, "a l'examen, necessites treure com a mínim un", round(nota_necessaria_exercicis, 2), "als exercicis.")
```

```python

euros = int(input("Introdueix una quantitat d'€: "))


bitllets = [5, 10, 20, 50, 100]

print(f"\nDescomposició de {euros} €:")


for valor in bitllets:
    quantitat = euros // valor
    sobra = euros % valor
    print(f"{quantitat} bitllets de {valor} i me'n sobren {sobra}")
```

```python
# Demanar quantitat d'euros
euros = int(input("Introdueix una quantitat d'euros: "))

# Bitllets de 100
bitllets_100 = euros // 100
restant = euros % 100

# Bitllets de 50
bitllets_50 = restant // 50
restant = restant % 50

# Bitllets de 20
bitllets_20 = restant // 20
restant = restant % 20

# Bitllets de 10
bitllets_10 = restant // 10
restant = restant % 10

# Bitllets de 5
bitllets_5 = restant // 5
restant = restant % 5

# Mostrar resultat
print(f"\n{euros} són:")
print(f"{bitllets_100} bitllets de 100")
print(f"{bitllets_50} bitllets de 50")
print(f"{bitllets_20} bitllets de 20")
print(f"{bitllets_10} bitllets de 10")
print(f"{bitllets_5} bitllets de 5")

# Si en sobren monedes no cobertes per bitllets
if restant > 0:
    print(f"I en sobren {restant} euros que no es poden descompondre amb bitllets.")
```
